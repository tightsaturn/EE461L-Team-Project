```python
page = requests.get('https://pokemondb.net/move/all')
soup = BeautifulSoup(page.content, 'html.parser')
results = soup.find(id='moves')
```

```python
movesTable = soup.find(id='moves')
moves = results.find_all('tr')
```

```python
for move in moves[1:]:    
    # navigate to move page
    move = move.find('a', class_='ent-name')
    moveURL = 'https://pokemondb.net' + move['href']
    movePage = requests.get(moveURL)
    moveSoup = BeautifulSoup(movePage.content, 'html.parser')
    
    # get move name
    moveName = move.text
    moveName = moveName.replace(' ', '-')
    print(moveName)
    
#     # get other data about move
#     tableSoup = moveSoup.find('table', class_='vitals-table')
#     tableSoup = tableSoup.find_all('td')
    
#     moveType = tableSoup[0].find('a', class_='type-icon').text
#     moveCategory = tableSoup[1].find('span', class_='img-fixed')['data-alt']
#     movePower = tableSoup[2].text
#     moveAccuracy = tableSoup[3].text
#     movePP = tableSoup[4].text
#     moveGen = tableSoup[5].text
    
#     effectSoup = moveSoup.find('div', class_="span-md-8")
#     moveEffect = moveSoup.find('p').text
    
    # get all the pokemon with the move
    pokemon = moveSoup.find_all('div', class_='infocard')
    pokeList = []
    pokeIDList = []
    for poke in pokemon:
        pokeName = poke.find('a', class_='ent-name').text
        pokeID = poke.find_all('small', class_='text-muted')
        for id in pokeID:
            if(id.text[0] == '#'):
                pokeID = int(id.text[1:4].strip('0'))
                break
                
        if(pokeIDList and pokeID == pokeIDList[-1]):
            continue
        
        pokeList.append(pokeName)
        pokeIDList.append(pokeID)
        
    # form json and add to database
    json = {
        'name': moveName,
        'pokemon': pokeList,
        'pokeID': pokeIDList
    }
        
    collection.insert_one(json)
    
```
